```python
import xarray as xr
import warnings
warnings.filterwarnings('ignore')
syr     = 1979
eyr     = 2000
neof		= 1
latS		= 20.
latN		= 90.
nyrs  = eyr - syr + 1

# Read Data ====================================================================
wdir="/data1/jhchoi/METRIC_NCL"

inDir = wdir+"/DATA/JRA55/hgt_mon/"
```

### Read Data
* python의 xarray는 ncl보다 array의 호환성과 확장성이 높음
* data를 부르고 바로 병합이 가능하기 때문에 비어있는 배열 생성 과정이 불필요함


```python
inFname=[]
for year in range(0,nyrs):
    inFname.append(f'{inDir}hgt.mon.{year+syr}.nc')

inFname=[f'{inDir}hgt.mon.{year+syr}.nc' for year in range(0,nyrs)]
```

```python
hgt=[]
for fname in inFname:
    inf=xr.open_dataset(fname)
    jhgt=inf['hgt']
    tmp=jhgt[:,jhgt['lev']==1000,:,:]
    hgt.append(tmp)

# jhgt=[xr.open_dataset(fname)['hgt'] for fname in inFname]
# hgt=[fname[:,fname['lev']==1000,:,:] for fname in jhgt]
hgt[-1]
```

```python
hgt=xr.concat(hgt, dim="time")
hgt
```

```python
hgt['time'].data
```

```python
hgt=hgt.rename({'lat':'latitude','lon':'longitude'})  
hgt=hgt.squeeze('lev')
hgt=hgt[:,latS<=hgt['latitude'],:]
hgt
```

```python
hgt2 = hgt.groupby('time.month').mean('time')
hgt2
```

```python
hgt_ano=hgt.groupby('time.month')-hgt2
hgt_ano
```
